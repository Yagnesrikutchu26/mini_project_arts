```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
shiva = pd.read_csv("handicrafts_traditional_arts_eda.csv")
shiva.head()
```

```python
shiva["Category"].value_counts()
```

```python
plt.figure()
shiva['Category'].value_counts().plot(kind='bar')
plt.title("shivamani arts")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
shiva['State'].value_counts()
```

```python
plt.figure()
shiva['State'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("State-wise Distribution of Handicrafts")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(
    x='Artisans_Involved',
    y='Craft_Name',
    data=shiva,
    estimator=sum
)
plt.title("Artisans_Involved per Craft")
plt.show()
```

```python
plt.figure()
sns.heatmap(shiva[ [  'Avg_Price_INR','Artisans_Involved','Annual_Production_Units'  ] ].corr (),
            annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
